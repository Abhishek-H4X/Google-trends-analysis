```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time
```

```python
Trending_topics = TrendReq(hl='en-US', tz=360)
kw_list = ["ChatGPT", "AI", "Cloud Computing", "Blockchain", "Cybersecurity"]
Trending_topics.build_payload(kw_list,cat=0,timeframe='today 12-m')
```

```python
data = Trending_topics.interest_over_time()
print(data)
```

```python
print(data[["ChatGPT", "AI"]].head(20))
```

```python
data = Trending_topics.interest_by_region()
data = data.sort_values(by="ChatGPT", 
                        ascending = False)
print(data["ChatGPT"].head(10))
```

```python
comparison = data[["ChatGPT", "AI"]]

print(comparison.head(10))
```

```python
print(data["ChatGPT"].mean())
print(data["AI"].mean())
```

```python
print(data["ChatGPT"].min())
```

```python
print(data[["ChatGPT","AI"]].idxmax())
```

```python
top10 = data.sort_values(by="ChatGPT", ascending=False).head(10)
top10["ChatGPT"].plot(kind="bar")
plt.show()
```

```python
data[["ChatGPT", "AI"]].plot(figsize=(12,6))

plt.title("ChatGPT vs AI Google Trends")
plt.xlabel("Date")
plt.ylabel("Interest")
plt.grid(True)

plt.show()
```
